```python
# Use this cell to begin your analysis, and add as many as you would like!
import requests
from collections import Counter
import re
from nltk.corpus import stopwords

# Download the text from Project Gutenberg
url = "https://www.gutenberg.org/files/16/16-0.txt"
response = requests.get(url)
text = response.text

# Extract the main text (remove header and footer)
start_marker = "*** START OF THE PROJECT GUTENBERG EBOOK PETER AND WENDY ***"
end_marker = "*** END OF THE PROJECT GUTENBERG EBOOK PETER AND WENDY ***"
start_index = text.find(start_marker) + len(start_marker)
end_index = text.find(end_marker)
main_text = text[start_index:end_index]

# Preprocess the text: lowercase, remove punctuation, split into words
words = re.findall(r'\b\w+\b', main_text.lower())

# Filter out stopwords
stop_words = set(stopwords.words('english'))
filtered_words = [word for word in words if word not in stop_words]

# Count word frequencies
word_counts = Counter(filtered_words)
top_words = [word for word, count in word_counts.most_common(10)]

# List of known character names from Peter and Wendy (lowercase for comparison)
character_names = {'peter', 'wendy', 'hook', 'john', 'michael', 'tinker', 'bell', 'smee', 'tink', 'nana', 'liza', 'curly', 'toodles', 'twin'}

# Check which of the top words are character names
protagonists = [word.capitalize() for word in top_words if word in character_names]

print(protagonists)
```
